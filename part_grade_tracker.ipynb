```python
# -----------------------------------------
# Task 1 — Data Parsing & Profile Cleaning
# -----------------------------------------

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []   # store cleaned data here

for student in raw_students:

    # --- Clean the name ---
    name = student["name"].strip().title()

    # --- Convert roll to integer ---
    roll = int(student["roll"])

    # --- Convert marks string to list of integers ---
    marks = [int(m) for m in student["marks_str"].split(", ")]

    # --- Validate name ---
    is_valid = all(word.isalpha() for word in name.split())

    if is_valid:
        print(f"{name} → ✓ Valid name")
    else:
        print(f"{name} → ✗ Invalid name")

    # Store cleaned student
    cleaned_students.append({
        "name": name,
        "roll": roll,
        "marks": marks
    })

    # --- Print profile card ---
    print("=" * 32)
    print(f"Student : {name}")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("=" * 32)
    print()

# -----------------------------------------
# Print name in ALL CAPS and lowercase for roll 103
# -----------------------------------------

for student in cleaned_students:
    if student["roll"] == 103:
        print("Special Output for Roll 103:")
        print("Name in UPPERCASE :", student["name"].upper())
        print("Name in lowercase :", student["name"].lower())
```

```python
# Student data
student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

# Function to get grade
def get_grade(m):
    if 90 <= m <= 100:
        return "A+"
    elif 80 <= m <= 89:
        return "A"
    elif 70 <= m <= 79:
        return "B"
    elif 60 <= m <= 69:
        return "C"
    else:
        return "F"

# Print subject, marks, and grade
print("Report Card for", student_name)
print()

for i in range(len(subjects)):
    print(subjects[i], ":", marks[i], "Grade:", get_grade(marks[i]))

# Total, average, highest, lowest
total = sum(marks)
average = round(total / len(marks), 2)

highest = max(marks)
lowest = min(marks)

highest_subject = subjects[marks.index(highest)]
lowest_subject = subjects[marks.index(lowest)]

print("\nTotal Marks:", total)
print("Average Marks:", average)
print("Highest Scoring:", highest_subject, highest)
print("Lowest Scoring:", lowest_subject, lowest)

# While loop to add new subjects
print("\nAdd new subjects (type 'done' to stop):")

new_count = 0

while True:
    sub = input("Enter subject name: ")

    if sub.lower() == "done":
        break

    mark_input = input("Enter marks: ")

    if not mark_input.isdigit():
        print("Invalid marks! Try again.")
        continue

    mark = int(mark_input)

    if mark < 0 or mark > 100:
        print("Marks must be between 0 and 100.")
        continue

    subjects.append(sub)
    marks.append(mark)
    new_count += 1

print("\nNew subjects added:", new_count)

updated_avg = round(sum(marks) / len(marks), 2)
print("Updated Average Marks:", updated_avg)
```

```python
# Class data
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print("Name              | Average | Status")
print("----------------------------------------")

averages = []
pass_count = 0
fail_count = 0

# Loop through each student
for name, marks in class_data:
    avg = round(sum(marks) / len(marks), 2)
    averages.append(avg)

    status = "Pass" if avg >= 60 else "Fail"

    if status == "Pass":
        pass_count += 1
    else:
        fail_count += 1

    print(f"{name:<17} | {avg:7.2f} | {status}")

# Find topper
top_avg = max(averages)
top_index = averages.index(top_avg)
topper_name = class_data[top_index][0]

# Class average
class_avg = round(sum(averages) / len(averages), 2)

print("\nSummary:")
print("Students Passed :", pass_count)
print("Students Failed :", fail_count)
print("Class Topper    :", topper_name, "-", top_avg)
print("Class Average   :", class_avg)
```

```python
# Given essay
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# 1. Strip leading and trailing whitespace
clean_essay = essay.strip()
print("1) Stripped Essay:")
print(clean_essay)
print()

# 2. Convert to Title Case
title_case = clean_essay.title()
print("2) Title Case Version:")
print(title_case)
print()

# 3. Count occurrences of "python" (case-insensitive)
count_python = clean_essay.count("python")
print("3) Count of 'python':", count_python)
print()

# 4. Replace "python" with "Python 🐍"
replaced = clean_essay.replace("python", "Python 🐍")
print("4) Replaced Version:")
print(replaced)
print()

# 5. Split into sentences
sentences = clean_essay.split(". ")
print("5) Sentences List:")
print(sentences)
print()

# 6. Numbered sentences with proper ending
print("6) Numbered Sentences:")
for i, sentence in enumerate(sentences, start=1):
    if not sentence.endswith("."):
        sentence += "."
    print(f"{i}. {sentence}")
```
